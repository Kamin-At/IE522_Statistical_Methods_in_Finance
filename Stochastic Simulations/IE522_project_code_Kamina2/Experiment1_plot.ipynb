```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('computing_time_comparison.csv')
actual_price_from_BSM = 5.07463
df
```

```python
plt.scatter(np.log10(df['n']), df['without_variance_reduction_price'], label='Without Antithetic', s=8)
plt.scatter(np.log10(df['n']), df['with_variance_reduction_price'], label='With Antithetic', s=8)
plt.plot(np.log10(df['n']), [actual_price_from_BSM] * len(df), '--', c ='r', label='Benchmark Price from BSM')
plt.legend(loc='lower right')#, bbox_to_anchor=(1, 0.75))
plt.grid()
plt.title('Comparison of output prices between use and not use Antithetic technique')
plt.ylabel('Price')
plt.xlabel('Log10(N)')
plt.show()
```

```python
plt.scatter(np.log10(df['n']), df['without_variance_reduction_standard_error'], label='Without Antithetic', s=8)
plt.scatter(np.log10(df['n']), df['with_variance_reduction_standard_error'], label='With Antithetic', s=8)
plt.legend(loc='upper right')#, bbox_to_anchor=(1, 0.5))
plt.grid()
plt.title('Comparison of standard errors between use and not use Antithetic technique')
plt.ylabel('Standard Error[$]')
plt.xlabel('Log10(N)')
plt.show()
```

```python
plt.scatter(np.log10(df['n']), np.log10(df['without_variance_reduction_time']), label='Without Antithetic', s=8)
plt.scatter(np.log10(df['n']), np.log10(df['with_variance_reduction_time']), label='With Antithetic', s=8)
plt.legend(loc='upper left')#, bbox_to_anchor=(1, 0.5))
plt.grid()
plt.title('Comparison of computing time between use and not use Antithetic technique')
plt.ylabel('Log10(Computing time[second])')
plt.xlabel('Log10(N)')
plt.show()
```
